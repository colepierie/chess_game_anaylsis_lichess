```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
csv = ['players.csv']

colum = ['game_id','format','player_id','player_rating','book_moves','result','points','rating_difference','color',
        'player_level']
```

```python
players = pd.DataFrame(columns = colum)

for file in csv:
    load_df = pd.read_csv(f'data/{file}')
    
    load_df = load_df[colum]
    
    players = players.append(load_df,ignore_index=True)
```

```python
white = players.loc[players["color"]=="white", :]
black = players.loc[players["color"]=="black", :]
```

```python
#create new category called 
outcome = []
for value in black['result']:
    if value == 'Lost':
        outcome.append('Loss')
    else:
        outcome.append('Win or Draw')
        
black['outcome'] = outcome
```

```python
X = black[['player_rating',"rating_difference",'book_moves']]

y = black["outcome"]
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
predictions = classifier.predict(X_test)
print(f"First 10 Predictions:   {predictions[:10]}")
print(f"First 10 Actual labels: {y_test[:10].tolist()}")
```

```python
black_pred = pd.DataFrame({"Prediction": predictions, "Actual": y_test}).reset_index(drop=True)
black_pred
```
